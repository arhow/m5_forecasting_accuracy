```python
import pandas as pd
import numpy as np
```

```python
sub1 = pd.read_csv('submission_v1.csv')
sub2 = pd.read_csv('ori_submission_v1.csv')
```

```python
sub1.head()
```

```python
sub2.head()
```

```python
sub1.set_index([['sub1']*sub1.index.shape[0], sub1.id], inplace=True)
```

```python
sub2.set_index([['sub2']*sub2.index.shape[0], sub2.id], inplace=True)
```

```python
sub_blend = pd.concat([sub1, sub2])
sub_blend = sub_blend.groupby(level=1).mean()
sub_blend['id'] = sub_blend.index
sub_blend.reset_index(drop=True, inplace=True)
```

```python
sub_blend.to_csv('blend_submission.csv', index=False)
```
